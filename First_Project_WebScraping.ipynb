## <h2 style="color: #008080;">Getting all football players names</h2>

```python
list_p_name=soup.find_all(class_="hauptlink")
```

```python
name=list_p_name[0]
```

```python
name.find("a").get("title")
```

```python
list_names=[]
for element in list_p_name:
    name=element.find("a").get("title")
    list_names.append(name)    
len(list_names)
```

## <h2 style="color: #008080;">Getting all clubs names</h2>

```python
list_p_clubs=soup.find_all(class_="zentriert no-border-rechts")
list_p_clubs[0]
```

```python
club=list_p_clubs[0]
club.find("a").get("title")
```

```python
list_clubs=[]
for element in list_p_clubs:
    club=element.find("a").get("title")
    list_clubs.append(club)
len(list_clubs)
```

## <h2 style="color: #008080;">Getting all ages</h2>

```python
list_p_ages=soup.find_all("td")
list_p_ages
```

```python
list_age=[]
for element in list_p_ages:
    age=element.get_text()
    list_age.append(age)
list_age=list_age[5::11]
list_age
```

## <h2 style="color: #008080;">Getting all countries</h2>

```python
list_p_countries= soup.find_all("img", class_="flaggenrahmen")
```

```python
list_countries=[]
for element in list_p_countries:
    country=element.get("alt")
    list_countries.append(country)
list_countries=list_countries[:41]    
```

## <h2 style="color: #008080;">Getting all injuries</h2>

```python
list_p_injuries=soup.find_all("td",class_="links")
list_p_injuries[0].get_text()
list_injuries=[]
for element in list_p_injuries:
    injury=element.get_text()
    list_injuries.append(injury)
list_injuries
```

## <h2 style="color: #008080;">Getting all market values</h2>

```python
list_p_value=soup.find_all("td",class_="rechts")
float(list_p_value[0].get_text()[:-8].replace(",","."))
```

```python
list_values=[]
for element in list_p_value:
    value=float(element.get_text()[:-8].replace(",","."))
    list_values.append(value)
list_values
```

## <h2 style="color: #008080;">Getting all innit dates</h2>

```python
list_p_indate=soup.find_all("td",class_="zentriert")
list_p_indate[8].get_text()

list_indate=[]
for element in list_p_indate:
    indate=element.get_text()
    list_indate.append(indate)
list_indate=list_indate[3::5]
list_indate
```

## <h2 style="color: #008080;">Getting all finish dates</h2>

```python
list_p_findate=soup.find_all("td",class_="zentriert")
list_p_findate

```

```python
list_findate=[]
for element in list_p_findate:
    findate=element.get_text()
    list_findate.append(findate)
list_findate=list_findate[4::5]
len(list_indate)
```
